# Setup Hi-C map for metainference simulations

```python
import sys
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
hic_data=np.loadtxt('../experimental_data/hic_chr5-31177kb-31207kb_200bp_coverage.dat')
```

```python
start=hic_data[:,0].min()
start
```

```python
end=hic_data[:,1].max()
end
```

```python
res=200 # bp
```

```python
L=int((end-start)/res+1)
L
```

```python
hic=np.zeros((L,L))
for loc1,loc2,h in hic_data:
    i=int((loc1-start)/res)
    j=int((loc2-start)/res)
    if(j>i):
        hic[i,j]=h
    elif(i>j):
        hic[j,i]=h
```

```python
fout=open('run_meta64_coverage13_all/hic_data_coverage.dat','w')
for i in range(L):
    for j in range(i+1,L):
        fout.write('%d %d %.1f\n'%(i+1,j+1,hic[i,j]))
fout.close()
```

```python
plt.imshow(np.log(hic+0.01))
```
