```python
dmaps=[]
cmaps_sigma13=[]
cmaps_sigma22=[]
for i in range(Nrep):
    for t in range(te-tb+1):
        dist=self_distance_array(trajs[i][t])
        dmaps.append( np.array(dist) )
        cmap=kernel_rational(dist,sigma=130.)
        cmaps_sigma13.append( np.array(cmap) )
        cmap=kernel_rational(dist,sigma=220.)
        cmaps_sigma22.append( np.array(cmap) )
dmaps=np.array(dmaps)
cmaps_sigma13=np.array(cmaps_sigma13)
cmaps_sigma22=np.array(cmaps_sigma22)
```

```python
def contact_scaling(hic):
        L=len(hic)
        scaling=np.zeros(L-1)
        obs    =np.zeros(L-1)
        for i in range(L):
                for j in range(i+1,L):
                        scaling[j-i-1]+=hic[i,j]
                        obs[j-i-1]    +=1.0
        scaling/=obs
        return scaling
```

```python
contacts_p1_all=[]
contacts_p2_all=[]
contacts_p3_all=[]
EPdistances=[]
for i in range(Nrep):
    for positions in trajs[i]:
        contacts_p1=np.zeros(L-1)
        contacts_p2=np.zeros(L-2)
        contacts_p3=np.zeros(L-3)
        for j in range(L-1):
            p1=positions[j]
            p2=positions[j+1]
            dist=np.sqrt(np.dot(p1-p2,p1-p2))
            contact=kernel_rational(dist,sigma=110.0)
            contacts_p1[j]=contact
            if j<L-2: # can compute i+2 contacts
                p2=positions[j+2]
                dist=np.sqrt(np.dot(p1-p2,p1-p2))
                contact=kernel_rational(dist,sigma=110.0)
                contacts_p2[j]=contact
            if j<L-3: # can compute i+3 contacts
                p2=positions[j+3]
                dist=np.sqrt(np.dot(p1-p2,p1-p2))
                contact=kernel_rational(dist,sigma=110.0)
                contacts_p3[j]=contact
        contacts_p1_all.append(np.array(contacts_p1))
        contacts_p2_all.append(np.array(contacts_p2))
        contacts_p3_all.append(np.array(contacts_p3))
        d=LA.norm(positions[i_pro]-positions[i_pro-25])
        EPdistances.append(d)
contacts_p1_all=np.array(contacts_p1_all)
contacts_p2_all=np.array(contacts_p2_all)
contacts_p3_all=np.array(contacts_p3_all)
EPdistances=np.array(EPdistances)
```
