## Locus statistsics

```python
sigma=22.0 # based on 1CPN simulations
```

```python
def radgyr(coordinates):
    com=coordinates.mean(axis=0)
    ri_sq = (coordinates-com)**2
    sq = np.sum(ri_sq, axis=1)
    rg = np.sqrt(sq.mean())
    return rg
```

```python
def traj2rg(trajs,tb=0,te=-1,skip=1):
    values=[]
    for u in trajs: # loop over the trajectories
        for ts in u.trajectory:
            if(ts.frame%skip!=0): continue
            if(ts.frame<tb): continue
            if(ts.frame>te and te>=0): break
            rg=radgyr(u.atoms.positions)
            values.append(rg)
    return np.array(values)
```

```python
# note, ids numbered from 1 to L
def traj2dist(trajs,id1,id2,tb=0,te=-1,skip=1):
    values=[]
    for u in trajs: # loop over the trajectories
        for ts in u.trajectory:
            if(ts.frame%skip!=0): continue
            if(ts.frame<tb): continue
            if(ts.frame>te and te>=0): break
            dist=u.atoms.positions[id1]-u.atoms.positions[id2]
            values.append(LA.norm(dist))
    return np.array(values)
```

## Autocorrelation

```python
# <s(t)s(t+tau)>t
def autocorr(trajs):
    trajs0=(trajs-trajs.mean())
    corrs=[]
    for traj in trajs0:
        result = np.correlate(traj,traj,mode='full')
        corrs.append( result[result.size//2:] ) # consider only values for positive time differences
        #corrs.append( result )
    corrs = np.array(corrs)
    acorr = corrs.mean(axis=0)
    acorr /= acorr.max()
    return acorr
```

## Diffusion

\begin{equation}
\langle({\bf r}(t+\tau)-{\bf r}(t))^2\rangle = \frac{t_{max}}{N_{traj}\tau} \sum_{i=1}^{N_{traj}}\sum_{t=0}^{t_{max}-\tau} ({\bf r}_i(t+\tau)-{\bf r}_i(t))^2
\end{equation}
In our case ${\bf r}={\bf r}_{E3}-{\bf r}_{E1}$

Compare to estimate given in:  
Bohrer, Christopher H., and Daniel R. Larson. "Synthetic analysis of chromatin tracing and live-cell imaging indicates pervasive spatial coupling between genes." Elife 12 (2023): e81861.
